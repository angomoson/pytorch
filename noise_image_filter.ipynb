```python
import cv2
import numpy as np
```

```python
# Load the noisy image
noisy_image = cv2.imread("/content/kim-ji-won-3.jpeg", cv2.IMREAD_GRAYSCALE)
```

```python
# Define the filter size (3x3 mean filter)
kernel_size = (3, 3)
```

```python
# Apply mean filter
denoised_image = cv2.blur(noisy_image, kernel_size)
```

```python
from google.colab.patches import cv2_imshow
```

```python
# Display the original and denoised images
cv2_imshow(noisy_image)
cv2_imshow(denoised_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
